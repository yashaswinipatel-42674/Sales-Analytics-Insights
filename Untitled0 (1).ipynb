```python
# Step 1: Upload file
from google.colab import files
uploaded = files.upload()

# Step 2: Import pandas
import pandas as pd

# Step 3: Read the Excel file (change name if your file is different)
df = pd.read_excel(list(uploaded.keys())[0])

# Step 4: Show first 5 rows
df.head()
```

```python
# Check structure of dataset
df.info()

# Quick statistics for numerical columns
df.describe()

# Check missing values
df.isnull().sum()
```

```python
# Remove duplicates
df = df.drop_duplicates()

# Convert Date column to proper format (if you have a Date column)
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Fill missing values in Customer Satisfaction with average
if 'Customer Satisfaction' in df.columns:
    df['Customer Satisfaction'].fillna(df['Customer Satisfaction'].mean(), inplace=True)

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x='Region', y='Revenue', data=df, estimator=sum)
plt.title("Revenue by Region")
plt.show()
```

```python
# Profit by Product Category
sns.barplot(x='Product Category', y='Profit', data=df, estimator=sum)
plt.title("Profit by Product Category")
plt.xticks(rotation=45)
plt.show()

# Customer Satisfaction distribution
sns.histplot(df['Customer Satisfaction'], bins=5, kde=True)
plt.title("Customer Satisfaction Distribution")
plt.show()
```

```python
# 1. Revenue trend over time
df.groupby('Date')['Revenue'].sum().plot(kind='line', figsize=(10,5), title="Revenue Over Time")

# 2. Top 5 products by revenue
top_products = df.groupby('Product Name')['Revenue'].sum().sort_values(ascending=False).head(5)
top_products.plot(kind='bar', title="Top 5 Products by Revenue")

# 3. Relationship between Profit and Revenue
sns.scatterplot(x='Revenue', y='Profit', data=df)
plt.title("Profit vs Revenue")
plt.show()
```

1. The Top 5 products contribute the largest share of total revenue .
2. Products with higher revenue generally show higher profit , indicating a positive correlation.
3. A few products generate high revenue but relatively low profit ,which may need cost optimization
4. Sales are not evenly distributed across products
